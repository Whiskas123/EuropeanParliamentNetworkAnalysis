```python
import networkx as nx
import pandas as pd


def generateGraph(country):
    edges = pd.read_csv(country + '/edges.csv')
    nodes = pd.read_csv(country + '/nodes.csv')
    G = nx.from_pandas_edgelist(edges,'Source','Target', edge_attr='Weight')

    nx.set_node_attributes(G, pd.Series(nodes['name'].values, index=nodes['Id']).to_dict(), 'Name')
    nx.set_node_attributes(G, pd.Series(nodes['group'].values, index=nodes['Id']).to_dict(), 'Group')
    nx.set_node_attributes(G, pd.Series(nodes['country'].values, index=nodes['Id']).to_dict(), 'Country')
            
    return G

def drawGraph(G):
    colors = []
    for node in G.nodes():
        if G.nodes[node]['Group'] == 'Verts/ALE':
            colors.append('green')
        elif G.nodes[node]['Group'] == 'S&D':
            colors.append('red')
        elif G.nodes[node]['Group'] == 'GUE/NGL':
            colors.append('purple')
        elif G.nodes[node]['Group'] == 'ECR':
            colors.append('darkblue')
        elif G.nodes[node]['Group'] == 'ENF':
            colors.append('black')
        elif G.nodes[node]['Group'] == 'ALDE':
            colors.append('yellow')
        elif G.nodes[node]['Group'] == 'NI':
            colors.append('grey')
        elif G.nodes[node]['Group'] == 'EPP':
            colors.append('blue')
        elif G.nodes[node]['Group'] == 'EFDD':
            colors.append('ligthblue')
    print(colors)
    nx.draw(G, node_color=colors)
```

```python
G = generateGraph('Austria')
```
